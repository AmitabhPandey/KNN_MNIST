```python
import numpy as np
import matplotlib.pyplot as plt

# Define dimensionalities: 2^0 to 2^10
log_dimensions = np.arange(11)
dimensions = 2 ** log_dimensions

means = []
standard_deviations = []

# Function to compute pairwise squared Euclidean distances
def compute_pairwise_distances(data):
    n = len(data)
    squared_euclidean_distances = []
    for i in range(n):
        for j in range(i + 1, n):
            squared_euclidean_distance = np.sum((data[i] - data[j]) ** 2)
            squared_euclidean_distances.append(squared_euclidean_distance)
    return squared_euclidean_distances

# Generate random samples and compute distances
for dim in dimensions:
    samples = np.random.rand(100, dim)  # 100 samples in 'dim'-dimensional space
    squared_euclidean_distances = compute_pairwise_distances(samples)
    means.append(np.mean(squared_euclidean_distances))
    standard_deviations.append(np.std(squared_euclidean_distances))

# Plotting results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(dimensions, means, marker='o', label="Mean")
plt.xlabel("Dimensions")
plt.ylabel("Mean Squared Distance")
plt.title("Mean vs Dimension")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(dimensions, standard_deviations, marker='s', color='orange', label="Std Dev")
plt.xlabel("Dimensions")
plt.ylabel("Standard Deviation")
plt.title("Std Dev vs Dimension")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
### Comparison of the above obtained Experimental value with the theoretical Values

theoretical_mean=dimensions/6

plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
plt.plot(dimensions, means, marker='o', label="Experimental Value", color='orange')
plt.plot(dimensions, theoretical_mean, marker='o', label="Theoretical Value",color='red')
plt.xlabel("Dimensions")
plt.ylabel("Mean")
plt.title("Theoretical Mean vs Experimental Mean")
plt.grid(True)
plt.legend()

theoretical_variance=np.sqrt((7*dimensions)/180)
plt.subplot(1,2,2)
plt.plot(dimensions, standard_deviations, marker='s', color='orange', label="Experimental Value")
plt.plot(dimensions, theoretical_variance, marker='s', color='red', label="Theoretical Value")
plt.xlabel("Dimensions")
plt.ylabel("Standard Deviation")
plt.title("Theoretical Variance vs Experimental Variance")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
